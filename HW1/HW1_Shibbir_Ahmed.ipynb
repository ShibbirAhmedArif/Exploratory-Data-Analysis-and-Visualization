## Question 3  ( 10 pts)
Write a function named **max_name** that, given a value of **AreaName** and a year between 1980 and 2013, returns the origin (value in **OdName**) with the maximum immigrants.

```python
def max_name(place, year):

    place_df = df[df['AreaName'] == place]

    year_column = int(year)

    max_row = place_df[year_column].idxmax()

    country_name = place_df.loc[max_row, 'OdName']

    return country_name

print("Country with the maximum immigrants:")
print(max_name('Asia', 2013))
print(max_name('Asia', 1980))
print(max_name('Europe', 2000))
```

## Question 4  ( 15 pts)
Write a function named **max_year** that, given one parameter of origin ( value in **OdName**), returns the year with the maximum immigrants.

```python
def max_immigrant_year(origin):

    selected_data = df[df['OdName'] == origin]

    if selected_data.empty:
        return "No data available for the particular origin"

    immigrant_years = selected_data.columns[7:]

    selected_data[immigrant_years] = selected_data[immigrant_years].apply(pd.to_numeric, errors='coerce')

    max_immigrant_year = selected_data[immigrant_years].idxmax(axis=1).iloc[0]

    return max_immigrant_year

result1 = max_immigrant_year('Afghanistan')
result2 = max_immigrant_year('Brazil')
print("Year with the maximum immigrants:")
print(result1)
print(result2)
```

## Question 5  ( 15 pts)
Write a function named **match_name** which searches a specified pattern in the values of **OdName** and returns all the matched records.

For example, set **'Austr'** as the parameter, this function will return the dataframe with the records of both **Austrialia** and **Austria**.

```python
def match_name(pattern):
    records = df[df['OdName'].str.contains(pattern, case=False)]
    return records

matched_df = match_name('Austr')
print(matched_df)
```
